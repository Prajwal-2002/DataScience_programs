```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.Class.unique()
```

```python
df2 = df.groupby(['Class'], as_index=False)['Petal_Width'].mean()
print(df2)
```

```python
plt.figure(figsize=[15,6])
sns.countplot(data=df, x='Sepal_Width', hue='Class', palette='Set1', saturation=0.9).set(title='Flowers of each species',xlabel='Sepal Width',ylabel='No. of flowers')
plt.tight_layout()
plt.show()
```

```python
intervals = (0, 1, 2, 4)
categories = ['<1', '1 to 2', '>2']
df['flowers_categories'] = pd.cut(df.Petal_Width, intervals, labels=categories)
sns.countplot(data=df, x='flowers_categories', hue='Class')
```

```python
plt.figure(figsize=[12, 6])
sns.countplot(data = df[df['Class']!='Iris-setosa'],x = 'Sepal_Width',hue='Class',palette='Set1').set(title='Versicolor vs virginica',xlabel='Sepal Width',ylabel='No. of flowers')
plt.tight_layout()
plt.show()
```
